```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from mpl_toolkits.axes_grid1 import inset_locator
```

```python
# in mm
steps = 100
diameter = np.arange(50,100,(100 - 50) / steps)
length = np.arange(10,3000,(3000 - 10) / steps)
```

```python
fig = plt.figure(figsize = (8,15))
ax = fig.add_subplot(311, projection = '3d')
ax1 = fig.add_subplot(312)
ax2 = fig.add_subplot(313)

#beam tubes
FIX_D = [60,80,100]

# surface plot 3D
x,y = np.meshgrid(diameter,length)
ax.plot_surface(X = x,Y = y,Z = 2*(np.arctan(x/2/y)))
ax.set_xlabel('diameter', labelpad = 10)
ax.set_ylabel('length', labelpad = 10)
ax.set_zlabel('kick_angle')

# sliced plot + zoom
for D in FIX_D:
    ax1.plot(length, 2*(np.arctan(D/2/length)))
    ax2.plot(length, length + 300/(np.tan(2*(np.arctan(D/2/length)))))
    
    
ax1.set_title('{} mm beam tube'.format(FIX_D))
ax1.set_xlabel('length')
ax1.set_ylabel('kick_angle [rad]')
ax1.legend(('60 mm', '80 mm', '100 mm'))

ax2.set_title('{} mm beam tube'.format(FIX_D))
ax2.set_xlabel('length')
ax2.set_ylabel('drift w/o optics')
ax2.legend(('60 mm', '80 mm', '100 mm'))
    


plt.savefig('/home/marius/Jlich/Multiplexer/kick_angle_eval')
plt.show()
```
